### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20            
hygiene_query = {'scores.Hygiene': {'$gt': 20}}
hygiene_sort = [('scores.Hygiene', -1)]

# Use count_documents to display the number of documents in the result
hygiene_full = list(establishments.find(hygiene_query).sort(hygiene_sort))
hygiene_first = establishments.find(hygiene_query).sort(hygiene_sort).limit(1)
pprint(len(hygiene_full))

# Display the first document in the results using pprint
pprint(list(hygiene_first))
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(hygiene_full)

# Display the number of rows in the DataFrame
print("Number of rows in result: ", len(hygiene_df))

# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
london_query = {'LocalAuthorityName': {'$regex': "London"},
                'RatingValue': {'$gt': 4}}
london_sort = [('RatingValue', -1)]
                
# Use count_documents to display the number of documents in the result
london_full = list(establishments.find(london_query).sort(london_sort))
london_first = establishments.find(london_query).sort(london_sort).limit(1)
pprint(len(london_full))
                
# Display the first document in the results using pprint
pprint(list(london_first))
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(london_full)

# Display the number of rows in the DataFrame
print("Number of rows in result: ", len(london_df))

# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
pprint(establishments.find_one({"BusinessName":'Penang Flavours'}))
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

#Base inputs
degree_search = 0.01
latitude = 51.490142 #hard coded to simplify 
longitude =  0.08384 #hard coded to simplify 
    
#Limits    
location_query = {'geocode.latitude': {'$lte' :(latitude + degree_search)},
                  'geocode.longitude': {'$lte' :(longitude + degree_search)},
                  'geocode.latitude': {'$gte' :(latitude - degree_search)},
                  'geocode.longitude': {'$gte' :(longitude - degree_search)},
                  'RatingValue': 5}    
location_sort = [('scores.Hygiene', -1)]

#Results
location_results = list(establishments.find(location_query).sort(location_sort).limit(5))

# Print the results
pprint(location_results)
                        
# Tried finding cleaner method, but only found examples that used greater than / less than functions. 
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
hygiene_pipe = {'$match': {'scores.Hygiene': 0}}
# 2. Groups the matches by Local Authority
group_pipe = {'$group': {'_id': "$LocalAuthorityCode", 'count': { '$sum': 1 }}}                
# 3. Sorts the matches from highest to lowest
sort_pipe = {'$sort': { 'count': -1 }}
                
hygiene_pipeline = [hygiene_pipe, group_pipe, sort_pipe]  
                
# Print the number of documents in the result
hyginen_group = list(establishments.aggregate(hygiene_pipeline))
                
# Print the first 10 results
pprint(hyginen_group[0:10])
```

```python
# Convert the result to a Pandas DataFrame
hyginen_group_df = pd.DataFrame(hyginen_group)
    
# Display the number of rows in the DataFrame
print("Number of rows in result: ", len(hyginen_group_df))

# Display the first 10 rows of the DataFrame
hyginen_group_df.head(10)
```
